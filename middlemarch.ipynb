```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import os.path
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python


# Set up Selenium with a WebDriver (e.g., ChromeDriver)
driver = webdriver.Chrome()

# Open the Project Gutenberg page for "Middlemarch"
url = "https://www.gutenberg.org/ebooks/145"
driver.get(url)

# Locate and click the link for the plain text file
link = driver.find_element("xpath", "//*[@id='download']/div/table/tbody/tr[8]/td[2]/a")

# driver.find_element(
link.click()
# //*[@id="download"]/div/table/tbody/tr[8]/td[2]/a

# Note: You might need to adjust the link text or element locator based on the specific page structure

# Wait for the file to download (time delay)
# You can implement a wait using WebDriverWait to ensure the download is complete
# Wait for the file to download (time delay)
# You can implement a wait using WebDriverWait to ensure the download is complete
wait = WebDriverWait(driver, 10)
wait.until(EC.url_matches(".txt"))


# Get the text from the downloaded plain text file
book_text = driver.page_source

# Close the browser
driver.quit()

# Use BeautifulSoup to parse the HTML and extract the text
soup = BeautifulSoup(book_text, "html.parser")
plain_text = soup.get_text()



save_path = 'data'

name_of_file = "middlemarch"

completeName = os.path.join(save_path, name_of_file+".txt")         

file = open(completeName, "w", encoding='utf-8')

file.write(plain_text)

driver.quit()


```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage you want to scrape
url = "https://www.litcharts.com/lit/middlemarch/characters"

# Send an HTTP GET request to the URL
response = requests.get(url)

character_list = []

if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    elements_with_name_class = soup.find_all(class_="name")

    # Iterate through the elements and print their content
    for element in elements_with_name_class:
        # print("Element with class='name':", element.text)
        character_list.append(element.text)

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
